```
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import pos_tag, wordpunct_tokenize, ne_chunk
from nltk.util import ngrams

import requests
from bs4 import BeautifulSoup

# Extracting from the Web URL

url = "https://en.wikipedia.org/wiki/Google"
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'html.parser')
data = soup.find('div', {'class': 'mw-parser-output'})
print(data.text)
with open('input.txt', 'w', encoding='utf-8') as f:
    f.write(str(data.text))

```

```
# Tokenization

nltk.download('punkt')
stokens = nltk.sent_tokenize(data.text)
print("Sentence Tokenization : \n", stokens)
printtokens = nltk.word_tokenize(data.text)

# Printing the word Tokenization
print ("Word Tokenization : \n", printtokens)
```

```
#Part Of Speech 

text = nltk.pos_tag(data.text)
print("Part Of Speech :")
print(text)
```
